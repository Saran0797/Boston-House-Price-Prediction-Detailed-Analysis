```python
def value_counts(data):
    for feature in data.columns:
        print (feature, data[feature].nunique())
```

```python
value_counts(continuous_feature)
```

```python
#to replace the extreme outliers with nan values

def extreme_outlier(data, threshold):
    for feature in data.columns:
        lower_quantile = data[feature].quantile(0.25)
        upper_quantile = data[feature].quantile(0.75)
        inter_quantile = upper_quantile-lower_quantile
        lower_boundary = lower_quantile-(threshold*inter_quantile)
        upper_boundary = upper_quantile+(threshold*inter_quantile)
        data.loc[data[feature]>upper_boundary, feature]=np.nan
        data.loc[data[feature]<lower_boundary, feature]=np.nan
```

```python
extreme_outlier(continuous_feature, 3)
```

```python
x = data.drop('price', axis=1)
```

```python
y = data.price
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=30)
```

```python
from sklearn.preprocessing import StandardScaler 
```

```python
std = StandardScaler()
std
```

```python
x_train = std.fit_transform(x_train)
pd.DataFrame(x_train, columns = ['CRIM', 'ZN', 'INDUS', 'NOX', 'RM', 'AGE', 'DIS', 'TAX', 'PTRATIO', 'B', 'LSTAT', 'CHAS', 'RAD'])
```

```python
x_test = pd.DataFrame(std.fit_transform(x_test), columns=['CRIM', 'ZN', 'INDUS', 'NOX', 'RM', 'AGE', 'DIS', 'TAX', 'PTRATIO', 'B', 'LSTAT', 'CHAS', 'RAD'])
x_test
```

```python
from sklearn.linear_model import LinearRegression
```

```python
linear = LinearRegression()
```

```python
linear.intercept_
```

```python
y_pred = linear.predict(x_test)
```
